- accuracy(정확도)  
    전체 데이터 중에 예측이 맞은 비율  
>   $accuracy = \frac{TP+TN}{TP+FP+TN+FN}$

- precision(정밀도)  
    양성으로 예측한 것 중에서 양성이 맞는 비율  
    즉, 예측한 거 중에 맞은 비율  
>   precision = \frac{TP}{TP+FP}$

- recall(재현률)  
    실제 양성인 것  중에서 양성으로 예측한 비율  
    즉, 맞은 비율  
>  $reacll = \frac{TP}{TP+FN}$

multi class 일 경우는, 각 클래스 별로 precision을 구하고 평균을 내면 최종 precision이 된다.  
최종 recall도 마찬가지 방법으로 구한다.

## Hyper Parameter

- 하이퍼 파라미터에 따라서 성능이 달라진다.  
고로, 최적의 하이퍼 파라미터를 찾는 것이 중요하다.
- 즉, 인간이 하이퍼파라미터(매개변수)를  넣어주고 컴퓨터가 최고의 파라미터(매개변수)를 찾는 것이다

````python
# 단순하게 하이퍼 파라미터를 for문으로 찾는 방법을 생각할 수 있다.
temp = []
for i in range(1, 20):  # 몇 개의 이웃을 보고 분류할 것인가? 1~19개로 설정 -> n_neighbors=1부터 n_neighbors=19까지 시도해보겠다는 뜻
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train, y_train)
    temp.append(knn.score(X_test, y_test))

# **테스트용 19개 아이리스 꽃들의 특징(X_test)**를 보고,
# 각각의 k값으로 라벨(꽃의 종류)을 예측해서,
# y_test와 비교했을 때 맞춘 비율 = 정확도가 리스트에 저장된 값들입니다.
````

### 교차검증(3way holdout)

데이터를 총 3등분한다.
- train set : 모델 학습에 사용한다.
- validationt set : 하이퍼 파라미터를 찾기 위해서 사용한다.
- test set : 한번도 안 본 데이터로 모델의 성능을 측정하기 위해서 사용한다.

````python
X_train, X_test, y_train, y_test = train_test_split(iris.iloc[:, :-1], iris.target)
````

````python
X_train, X_validation, y_train, y_validation = train_test_split(X_train, y_train)
````

 **"하이퍼파라미터 튜닝"**을 위한 3-way holdout 방식을 사용해서 최적의 K 값을 찾는 과정

K=1

````python
# 하이퍼 파라미터를 찾기 위해서 3-way holdout을 한다.
temp = []
for i in range(1, 20):
    knn = KNeighborsClassifier(n_neighbors=i) # 이웃의 수K를 보고 검사를 한다 K=1 (이웃1개만 보고 검사한다)
    knn.fit(X_train, y_train)
    temp.append(knn.score(X_validation, y_validation))
````

````python
temp
````

**정확도 94.7%**를 기록했다는 뜻

````python
# 3way holdout으로 했을 때, 하이퍼 파라미터 n_neighbors가 6일때 가장 성능이 좋다.
# temp[5] = K=6일 때 정확도 = **0.973 (최고)**
np.argmax(temp) # temp의 가장 성능이 좋은 인데스가 출력됨
````

#### cross_val_score
- 사이킷런(Scikit_Learn)에서는 교차검증(K-Fold, StratifiedKFold)을 쉽게 할수 있도록 제공하는 API 이다.
- scikit-learn 0.22 버전부터 기본적으로 5-겹 교차 검증으로 바뀌었다.
- cross_val_score에서 사용되는 파라미터
  - estimator : 분류(Classifier) 또는 회귀(Regressor)인지 구분
  - X : feature  데이터 세트
  - y : class 데이터 세트
  - scoring : 예측 성능 평가 지표('accuracy', 'neg_brier_score', 'top_k_accuracy'...)
  - cv : 교차 검증 필드 수
- y가바이너리 또는 다중 클래스, StratifiedKFold가 사용되며, 전체적으로 그 외의 경우 KFold 사용된다.    
- 수행 후 반환값은  scoring 파라미터로 지정된 측정값을 배열 형태로 반환한다.


````python
# cross_val_score로 하이퍼 파라미터 찾기
# 이 방법은 모든 데이터를 사용하기 때문에 위의 temp 방식 보다 많이 사용한다.
temp2 = [] # K값에 따른 평균 정확도를 저장

for i in range(1, 20):
    temp2.append(
        cross_val_score(
            KNeighborsClassifier(n_neighbors=i),
            iris.iloc[:, :-1],     # 특징 (X)
            iris.target,           # 레이블 (y)
            cv=5                   # 5겹 교차 검증
        ).mean()                   # 5개의 점수 평균을 저장
    )


#  교차 검증 분할 수 (`cv=5`이면 5-fold cross validation)
# 기본값은 `cv=5` (scikit-learn 0.22부터)
# `StratifiedKFold`은 분류 문제에서 클래스 비율을 **고르게 유지**
````

````python
from sklearn.linear_model import LogisticRegression
````

````python
from sklearn.model_selection import GridSearchCV
````

````python
# 전문 기업들은 전처리만 하고 나면, 자동화된 코드를 모두 만들어두었다.
# 함수로 만들어두면 재사용이 가능하다.

def all_models(X_train, y_train,  cv=10):
    ###################################################
    # 1. Logistic Regression                          #
    ###################################################
    param_grid_lr = {
        'C': [0.01, 0.1, 1, 10],
        'solver': ['liblinear'],  # 작은 데이터셋에서 안정적
        'penalty': ['l1', 'l2']
    }
    grid_lr = GridSearchCV(LogisticRegression(),param_grid_lr, cv=cv)
    grid_lr.fit(X_train, y_train)
    lr_pd = pd.DataFrame(grid_lr.cv_results_)
    ###################################################
    # 2. KNeighbors                                   #
    ###################################################
    param_grid_knn = {
        'n_neighbors': [3, 5, 7],
        'weights': ['uniform', 'distance'],
        'metric': ['euclidean', 'manhattan']
    }
    grid_knn = GridSearchCV(KNeighborsClassifier(), param_grid_knn , cv=cv)
    grid_knn.fit(X_train, y_train)
    knn_pd = pd.DataFrame(grid_knn.cv_results_)

    return pd.concat([lr_pd, knn_pd], axis=1)
````

# LogisticRegression().get_params()
-  LogisticRegression 모델의 모든 하이퍼파라미터와 그 기본값을 확인하는 코드입니다.

- **`LogisticRegression()`**: 로지스틱 회귀 모델의 인스턴스를 생성합니다.
- **`.get_params()`**: 이 메서드는 모델이 가지고 있는 모든 파라미터(하이퍼파라미터 포함)를 딕셔너리 형태로 반환합니다.

이 코드를 실행하면 다음과 같은 정보를 얻을 수 있습니다:

- **어떤 하이퍼파라미터를 조절할 수 있는지**: 모델의 성능을 튜닝하기 위해 변경할 수 있는 모든 하이퍼파라미터의 목록을 보여줍니다.
- **각 하이퍼파라미터의 기본값은 무엇인지**: 사용자가 특별히 값을 지정하지 않았을 때 어떤 값이 기본으로 사용되는지 알 수 있습니다.

`GridSearchCV`와 같은 도구를 사용하여 최적의 하이퍼파라미터를 찾을 때, 이 `get_params()`를 통해 확인한 파라미터 이름을 `param_grid`에 사용하게 됩니다.

## ③ GridSearchCV + Pipeline
(Pipeline은 전체 머신러닝 작업 흐름(전처리 → 모델 학습 등)을 한 덩어리로 묶어주는 도구)

## <GridSearchCV와 Pipeline을 연동하기>
- Pipeline과 GridSearchCV를 연동하면,
- 전처리부터 모델 학습까지의 전체 과정을
- 자동화 + 튜닝 + 교차검증까지 한 번에 처리할 수 있기 때문이에요.

### 쉽게설명
- 예를 들어, 여러분이 꽃의 품종을 분류하는 모델을 만든다고 해요.
- 그런데 데이터에는 스케일 차이가 큰 특성들이 섞여 있어요.
- 그래서 StandardScaler()로 스케일링을 먼저 하고
- KNN, SVM 같은 모델을 학습해야 해요.

### 이런 과정이 보통 이렇게 나뉘죠:

- 1. 데이터 스케일링 (StandardScaler)

- 2. 모델 선택 및 학습 (KNN)

- 3. 하이퍼파라미터 튜닝 (GridSearchCV)

- 4.  교차검증

### 그런데 문제는?
- 하이퍼파라미터 튜닝을 위해 GridSearchCV를 쓰면,
데이터를 나눠서 여러 번 훈련하고 평가해야 해요.

- 그런데 그때마다 스케일링도 반드시 훈련데이터에 맞춰 다시 해야 해요!

- 스케일러를 전체 데이터에 fit 한 후 나누면 → 데이터 누수(data leakage) 발생!

-  그래서 Pipeline이 필요!

> * 데이터 누수란?,
- AI가 시험(테스트 데이터)을 보기 전에 미리 답(정보)을 슬쩍 본 상태를 말해요!**
- *그래서 공정한 평가를 위해, 기준(fit)은 훈련 데이터로만!**



### ✏️ `fit()` = 공부/훈련 (AI가 훈련 데이터로 공부함)

- 이건 **훈련 데이터를 보면서**,
- **평균이 얼마고, 얼마나 퍼져 있는지를 기억하는 과정**이에요.

예:

“나는 지난번 수학 시험에서 평균 60점이었어. 다음엔 70점 맞아야지!”

---

### 📝 `transform()` = 테스트

- 이제 ****시험 문제(테스트 데이터)****가 나왔어요.
- 공부한 내용을 바탕으로 **시험을 공정하게 봐야 해요.**
- 시험지는 새롭지만, **공부할 때 배운 기준(평균 등)**만 적용해서 푸는 거예요.

---

## ❗ 데이터 누수 = 시험 전에 시험 문제를 미리 본 것!

만약 우리가 **시험지를 미리 보고 공부 기준을 세웠다면?**

- 시험지를 보고 “아~ 이건 자주 나오는 문제네!” 하면서
- **공부 기준 자체를 바꿔버렸다면?**
- 그건 **공정한 시험이 아니죠!!** → 즉, **컨닝이에요!**

### GridSearchCV로 최적의 알고리즘 찾기

`GridSearchCV`는 최적의 하이퍼 파라미터는 찾을 수 있지만, 최적의 알고리즘을 찾는 기능은 공식적으로 제공하지 않는다.  
그러나, scikit-learn의 고수들이 알고리즘도 찾아주는 꼼수를 발견하였다.

````python
pipe = Pipeline([('ss', StandardScaler()), ('clf', KNeighborsClassifier())])
````

````python
# pipe에서 clf를 하이퍼 파라미터로 취급하여, 알고리즘을 바꾼다.
grid = GridSearchCV(pipe, {'clf':[KNeighborsClassifier(), LogisticRegression()]})
````

````python
grid.fit(wine.iloc[:, :-1], wine.target)
````

````python
# 알고리즘을 바꿔가면서 GridSearchCV하게 만들 수 있다.
pd.DataFrame(grid.cv_results_)
````

### 알고리즘마다 하이퍼 파라미터 찾기

````python
# param_grid를 list 안의 dict로 넣으면, dictionary끼리만 조합을 만든다.
# 최적의 알고리즘과 최적의 하이퍼 파라미터를 동시에 찾을 수 있다.
grid = GridSearchCV(pipe, [{'clf':[KNeighborsClassifier()],
                            'clf__n_neighbors':[2,3,4,5]},
                           {'clf':[LogisticRegression()]}])
````

````python
# knn의 하이퍼 파라미터는 knn에서만 사용되었다.
pd.DataFrame(grid.cv_results_).T
````
