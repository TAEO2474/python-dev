```python
# pandas에서는 dictionary 방식 과 arrtibute 방식 모두 가능하다.
# dictionary 방식
data['tip'][0]  # data[열이름][행이름]

# arrtibute 방식
data.tip[0]     # data.열이름[행이름]
```

at
- 이름으로 한 위치의 값만 가져온다.

```python
print(data.at[0,'tip'])     # 1.01
print(data.iloc[0, 1])       # 1.01
print(data.loc[0,'tip'])    # 1.01
print(data['tip'][0])       # 1.01
```

```python
# at에서는 slicing을 사용할 수 없다.
print(data.at[0:3,'tip'])
```

```python
# at에서는 fancy indexing을 사용 할 수 없다.
print(data.at[[0,1,3],'tip'])
```

iat
- iat은 번호로 한 위치의 값을 가져온다.
- at와 마찬가지로 slicing과 fancy indexing을 사용할 수 없다.

```python
data.iat[0, 1]  #1.01
```

boolean indexing

```python
# data['tip'] >= 5
data[data['tip'] >= 5]
```

```python
data[data['tip'] >= 5]['tip']
```

```python
#data[data['tip'] >= 5].loc[:, 'tip':'day'] #'tip'부터 'day'까지
data[data['tip'] >= 5].iloc[:, 1:5]   #1번호부터 5미만
```
