## 정규 표현식 (정규식)
- 쉽게말해 원하는 문자(키워드) 찾는 도구
  - 복잡한 문자열을 처리할 때 사용하는 기법으로, 파이썬만의 고유 문법이 아니라 문자열을 처리하는 모든 곳에서 사용된다.   
  - 일반적인 문자(a~z)와 메타문자(meta characters)라고 부르는 특수한 문자를 조합해서 특정 문자열 패턴을 작성하기 위한 언어이다.

  - 메타문자 : 원래 그 문자가 가진 뜻이 아닌 특별한 용도로 사용되는 문자이다. 정규 표현식에서 사용되는 문자 . ^ $ * + ? {} [] \ | ()
  - 공식문서 : https://docs.python.org/3/howto/regex.html


  '정규 표현식'이라는 이름은 컴퓨터 과학의 이론 분야인 '형식 언어 이론(formal language theory)'에서 유래했습니다.

이 이론에서 '정규 언어(regular language)'라는 개념이 있는데, 이는 특정 규칙에 따라 생성될 수 있는 문자열의 집합을 의미합니다. '정규 표현식'은 이러한 '정규 언어'를 설명하거나 정의하는 데 사용되는 표기법입니다.

따라서, 정규 표현식은 특정 패턴에 맞는 "정규적인" 문자열들을 표현하기 때문에 '정규 표현식'이라고 불리게 되었습니다. 복잡해 보일 수 있지만, 기본적으로는 특정 규칙을 가진 문자열 패턴을 정의하는 도구라고 생각하시면 됩니다.

# <정규화 표현식 중요도 순위>
| 순위 | 문법        | 설명                            | 예시                                     |       |                    |
| -- | --------- | ----------------------------- | -------------------------------------- | ----- | ------------------ |
| 1  | `.` (dot) | 임의의 문자 1개와 매칭 (줄바꿈 제외)        | `a.c` → `abc`, `a9c`                   |       |                    |
| 2  | `*`       | 앞 문자가 0번 이상 반복                | `ab*c` → `ac`, `abc`, `abbc`           |       |                    |
| 3  | `+`       | 앞 문자가 1번 이상 반복                | `ab+c` → `abc`, `abbc` (하지만 `ac`는 불일치) |       |                    |
| 4  | `?`       | 앞 문자가 0번 또는 1번                | `ab?c` → `ac` 또는 `abc`                 |       |                    |
| 5  | `[]`      | 문자 집합 (안에 있는 문자 중 하나)         | `[abc]` → `a` 또는 `b` 또는 `c`            |       |                    |
| 6  | `^`       | 문자열 시작 지점 지정                  | `^a` → `a`로 시작하는 문자열                   |       |                    |
| 7  | `$`       | 문자열 끝 지점 지정                   | `a$` → `a`로 끝나는 문자열                    |       |                    |
| 8  | `()`      | 그룹핑 및 캡처                      | `(abc)+` → `abc`, `abcabc` 반복 매칭       |       |                    |
| 9  | `\d`      | 숫자(0\~9)와 매칭                  | `\d{3}` → 숫자 3개 연속                     |       |                    |
| 10 | `\D`      | 숫자가 아닌 문자와 매칭                 | `\D+` → 숫자 아닌 문자 여러 개                  |       |                    |
| 11 | `\s`      | 공백 문자 (스페이스, 탭, 줄바꿈)          | `\s+` → 공백 1개 이상                       |       |                    |
| 12 | `\S`      | 공백이 아닌 문자                     | `\S+` → 공백 아닌 문자 여러 개                  |       |                    |
| 13 | `\w`      | 단어 문자(알파벳, 숫자, 언더스코어)         | `\w+` → 단어 하나                          |       |                    |
| 14 | `\W`      | 단어 문자가 아닌 문자                  | `\W+` → 단어 문자가 아닌 문자들                  |       |                    |
| 15 | `{n}`     | 앞 문자가 정확히 n번 반복               | `\d{3}` → 숫자 3개 연속                     |       |                    |
| 16 | `{n,m}`   | 앞 문자가 n\~m번 반복                | `a{2,4}` → `aa`, `aaa`, `aaaa`         |       |                    |
| 17 | \`        | \`                            | OR 연산자 (둘 중 하나 선택)                     | \`cat | dog`→`cat`또는`dog\` |
| 18 | `\b`      | 단어 경계 (단어 시작/끝 위치)            | `\bword\b` → 정확한 ‘word’ 단어 찾기          |       |                    |
| 19 | `\\`      | 역슬래시 문자 자체                    | `\\` → 문자열 내에서 '' 문자 찾기                |       |                    |
| 20 | `\`       | 이스케이프 문자: 특수문자 문자화 또는 특수기호 표현 | `\.` (점 문자), `\n` (줄바꿈), `\t` (탭) 등    |       |                    |



```python
import re


pattern = r'비'
text = "하늘에 비가오고있습니다.어제도비가왔고 오늘도비가오고있습니다"
regex = re.compile(pattern) #패턴을 컴파일시킴
result = regex.findall(text)   #matiching이 이루어진 모든 문자열의리스트를반환
print(result)




```

## 정규식을 이용한 문자열 검색

### 점(.) 기호 - 한 개의 임의의 문자를 나타낸다.






```python
r = re.compile('a.c') # 'a'와 'c' 사이에 어떤 문자든 하나가 오는 패턴을 컴파일합니다.
```

```python
m = r.search('kabc') # 문자열 'kabc'에서 컴파일된 패턴과 일치하는 부분을 검색합니다.
print(m) # 매치된 객체를 출력합니다.
print(dir(m)) # 매치 객체의 속성과 메소드를 출력합니다.
print(m.group()) # 매치된 문자열 자체를 출력합니다 ('abc').
print(m.start()) # 매치된 문자열의 시작 인덱스를 출력합니다 (1).
print(m.end()) # 매치된 문자열의 끝 인덱스를 출력합니다 (4).
print(m.span()) # 매치된 문자열의 시작과 끝 인덱스를 튜플로 출력합니다 (1, 4).
```

```python
m = r.search('kabcaec', pos=5)
print(m)   # 정규식와 일치하는 데이터가 없으면 None을 리턴함
print(dir(m))
#print(m.group())
#print(m.start())
#print(m.end())
#print(m.span())
```

```python
m = r.search('kabcaec', pos=4)
print(m)
print(dir(m))
print(m.group())
print(m.start())
print(m.end())
print(m.span())
```

## 컴파일 옵션

|옵션명|약어|설명|
|---|---|---|

### DOTALL : 줄바꿈 문자를 포함하여 모든 문자와 매치할 수 있도록 한다.

```python
r = re.compile('a.b')
m = r.search('a\nb')
print(m)
```

```python
#r = re.compile('a.b', re.DOTALL)
r = re.compile('a.b', re.S)
m = r.search('a\nb')
print(m)
```

### IGNORECASE : 대소문자에 관계없이 매치할 수 있도록 한다.

```python
r = re.compile('[a-z]')
m = r.search('Life is to')
print(m)
```

```python
#r = re.compile('[a-z]', re.IGNORECASE)
r = re.compile('[a-z]', re.I)  # 대문자 아이
m = r.search('Life is to')
print(m)
```

### MULTILINE : 여러 라인에서 매치되는 데이터를 가져올 때 사용한다.

- \d : [0-9]와 동일한 표현식  
- \D : [^0-9]와 동일한 표현식
- \s : 공백 [\t\n\\r\f] 와 동일한 표현식
- \S : [^\t\n\\r\f] 와 동일한 표현식
- \w : [a-zA-Z0-9_]와 동일한 표현식
- \W : [^a-zA-Z0-9_]와 동일한 표현식

```python
#r = re.compile('^python\s\w+', re.MULTILINE)
r = re.compile('^python\s\w+', re.M)

# findall () : 패턴과 일치하는 부분 중에서 괄호 안 그룹만 리스트로 출력
m = r.findall('''python one
life is too short
python two
you need python
python _
python 9
python three''')


print(m)

```

```python
p = re.compile("&#(0[0-7]+" # 8진수 숫자 (0으로 시작하고 뒤에 0~7 숫자가 한 개 이상)
                     "|[0-9]+" # 10진수 숫자 (0~9 숫자가 한 개 이상)
                     "|x[0-9a-fA-F]+);")#16진수 숫자 (x 뒤에 09 또는 af/A~F 숫자가 한 개 이상)


data="""&#06;
&#9;
&#xa;"""

#  -> &#06; (8진수 '06')
# -> &#9; (10진수 '9')
# -> &#xa; (16진수 xa, 즉 0xa = '10')


# findall () : 패턴과 일치하는 부분 중에서 괄호 안 그룹만 리스트로 출력
m=p.findall(data)

print(m)
```

## 백슬래시 (\\) 문제

```python
#s을 whitespace로 해석을 해서 의도가 다르게 처리된다.
r = re.compile('\section') # \d  \w   \s

m = r.match(' ection')
print(m)
```

```python
# 백슬래시(\)가 문자열 의미로 정규 표현식에서 사용이 될때에는 파이썬 정규식 엔진에서는  \\\\4개를 명시를 해줘야한다.
r = re.compile('\\\\section') # \d  \w   \s

m = r.match('\\section')
print(m)
```

```python
r = re.compile(r'\\section') # \d  \w   \s

m = r.match(r'\section')  # raw string(일반 문자열)
print(m)
```

## 문자열 소모가 없는(zero-width assertions)메타 문자
| 메타문자는 or과 동일한 의미로 사용한다.

```python
p = re.compile('Crow|Servo')
m = p.match('CrowHello')
print(m)
print(m.group())
print(m.start())
print(m.end())
print(m.span())
```

# ^ 메타 문자는 문자열의 맨 처음과 일치함을 의미한다. re.MULTILINE을 사용할 경우에는 여러 줄의 문자열일 때 각 라인의 처음과 일치하게 된다.

```python
r=re.compile('^Life')
m=r.search('Life is too short')
print(m)
print(m.group())
print(m.start())
print(m.end())
print(m.span())
```

## 그룹핑
문자열이 계속해서 반복되는지 조사하는 정규식

####  (  )   : 그룹을 만들어 주는 메타문자이다.

|group(인덱스) | 설명 |
|---|---|
| group(0) | 매치된 전체 문자열|
| group(1) | 첫 번째 그룹에 해당되는 문자열 |
| group(2) | 두 번째 그룹에 해당되는 문자열 |
| group(n) | n 번째 그룹에 해당되는 문자열 |

```python
p=re.compile('(ABC)+')

m=p.search('ABCABCABC OK!')

print(m)
print(m.group(0), m.start(), m.end(), m.span())

print(m.group(1),  m.start(), m.end(), m.span())
```

```python
r = re.compile(r'\w+\s+(\d+[-]?){3}')

m = r.search('park 010-1234-5678')
print(m)
```

```python
# p = re.compile(r'(\w+)\s+((\d+)[-]\d+[-]\d+)')

p = re.compile(r'(\w+)\s+((\d+)-\d+-\d+)') # 그룹핑 인덱스 : 왼쪽-> 오른쪽,   바깥에서 -> 안으로

m = p.search('park 010-1234-5678')

print(m)

print(m.group(0), m.start(), m.end(), m.span())  # 전체

print(m.group(1),  m.start(), m.end(), m.span()) # (\w+)

print(m.group(2),  m.start(), m.end(), m.span()) #((\d+)-\d+-\d+)

print(m.group(3),  m.start(), m.end(), m.span()) #(\d+)
```

### 전방탐색(Lookahead Assertions)
전방 탐색에는 긍정(Positive)과 부정(Negative)의 2종류가 있고 다음과 같이 표현한다.


- 긍정형 전방 탐색((?=...)) - ... 에 해당되는 정규식과 매치되어야 하며 조건이 통과되어도 문자열이 소비되지 않는다.
- 부정형 전방 탐색((?!...)) - ...에 해당되는 정규식과 매치되지 않아야 하며 조건이 통과되어도 문자열이 소비되지 않는다.


```python
import re # re 모듈을 임포트합니다. 이 모듈은 파이썬에서 정규 표현식 작업을 할 수 있게 해줍니다.
r = re.compile('.+:') # '.+:' 패턴을 컴파일하여 정규 표현식 객체 'r'을 생성합니다.
                     # '.': 줄바꿈 문자를 제외한 모든 문자와 매치합니다.
                     # '+': 앞선 문자('.')가 한 번 이상 반복되는 것과 매치합니다.
                     # ':': 콜론 문자 그대로와 매치합니다.
                     # 따라서 이 패턴은 하나 이상의 문자가 오고 그 뒤에 콜론이 오는 문자열을 찾습니다.

m = r.search('http://google.com') # 문자열 'http://google.com' 전체를 검색하여 컴파일된 패턴 'r'과 일치하는 첫 번째 부분을 찾습니다.
                                  # 매치되는 부분이 있으면 Match 객체를 반환하고, 없으면 None을 반환합니다. 이 경우 'http:'와 매치됩니다.

print(m) # search() 메소드로부터 반환된 Match 객체를 출력합니다.
```

```python
import re # re 모듈을 임포트합니다.
r = re.compile('.+(?=:)') # '.+(?=:)' 패턴을 컴파일(찾기)
                         # '.+':“줄바꿈을 제외한 아무 글자나 하나 이상”즉, 글자가 여러 개 있는 걸 찾는다는 뜻이에요.
                         # '(?=:)': 이건 **"바로 뒤에 ':'가 있어야 한다"**는 조건이에요. 단, 그 콜론 :은 결과에는 포함하지 말라! (전방 긍정 탐색).
                         # 따라서 이 패턴은 콜론 앞에 오는 모든 문자를 찾습니다.

m = r.search('http://google.com') # 문자열 'http://google.com'에서 패턴과 일치하는 첫 번째 부분을 검색합니다. 이 경우 'http'와 매치됩니다.
print(m) # 매치된 Match 객체를 출력합니다.
```

```python
import re
r = re.compile('(?=:).+')
m = r.search('http://google.com')
print(m)
```

#### 부정적 전방탐색 ?!

```python
#p = re.compile('.*[.](?=bar$|cf$)')
p = re.compile('.*[.](?!bar$|cf$)') # .* = 아무 글자나 쭉쭉쭉 다!"(줄바꿈 제외, 가능한 한 많이)
                                    # [.] =  "그냥 점" 찾고,
                                    # (?=bar$|cf$) =  전방 긍정 탐색 (Lookahead)
file = ['foo.bar', 'autoexec.bat', 'sendmail.cf', 'sample.jpg' , 'test.bar']

for item  in file:
  m = p.search(item)
  #print(m)

  if m :
    print(m.group(), m.start(), m.end())
```

### re.sub()
- 정규 표현식 패턴과 일치하는 문자열을 찿아 다른 문자열로 대체하는 함수

# re.sub(pattern, replacement, string)
- pattern : 찾고 싶은 정규 표현식 패턴

- replacement : 바꿀 문자열

- string : 대상 문자열

```python
p=re.compile('(blue|white|red)')
#sub('바꿀 문자열', '대상문자열')
m=p.sub('color', 'blue socks and red shoes')
print(m)
```

### subn( )
- 정규표현식 문자열이 있으면 대체 문자열로 변경후 튜플로 리턴한다.

```python
p = re.compile('(blue|white|red)')
m = p.subn('color', 'blue socks and red socks')  # subn('대체문자', '대상문자')
print(m)
```

#### NLTK에서 정규 표현식을 사용한 단어 토큰화

```python
import nltk
from nltk.tokenize import RegexpTokenizer
```

```python
text = "Don't be fooled by the dark sounding name, Jone's Orphange is as cheery as goes for a pastry shop."
tokenizer = RegexpTokenizer( "[\w]+" )
print( tokenizer.tokenize( text ) )
```

```python
tokenizer = RegexpTokenizer( "[\s]+", gaps = True )
print( tokenizer.tokenize( text ) )
```

# RegexpTokenizer(r'\w+')  
 - 알파벳, 숫자, 밑줄로 이루어진 연속된 단어를 토큰으로 분리하는 정규 표현식 기반의 토큰화 함수입니다. 구두점은 제외하고, 단어만 추출하고 싶을 때 유용하게 사용됩니다

```python
import nltk
from nltk.tokenize import RegexpTokenizer

# 정규 표현식을 사용하여 토큰화할 패턴 정의
#  RegexpTokenizer() 함수 = 정규 표현식 패턴을 이용해 텍스트에서 **단어(토큰)**을 찾아내는 함수
#(r'\w+') = "알파벳, 숫자, 밑줄로 이루어진 단어"**를 하나 이상 찾아낸다는 뜻
tokenizer = RegexpTokenizer(r'\w+')

# 예시 문장
text = "I love AI! It's amazing."

# 토큰화 실행
tokens = tokenizer.tokenize(text)
print(tokens)
```
