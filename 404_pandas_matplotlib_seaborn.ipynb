### 데이터 시각화
- 데이터 시각화의 중요성을 강조하고자 할때 영국의 통계학자 프랭크 엔스컴이 만든 앤스컴 콰르텟(앤스컴 콰르텟:Ansombe's quartet)으로 데이터를 시각화하지 않고 수치만 확인할 때 발생할 수 있는 함정을 보여주기 위한 그래프이다.
- 앤 스컴 콰르텟 은 4 개의 데이터 셋 으로 구성 되며 각 데이터 셋 에는 2 개의 연속 변수 가 있 습니다 . 4 개 데이터 셋 은 평균 , 분산 , 상관 관계 , 회귀선 이 모두 같다 . 따라서 기술 통계 만 보면마치 같은 데이터 셋 처럼 보일 수 있다 . 하지만 이를 시각화 하면 4 가지 모두 경향 이 다르다는 사실 을 직관적 으로 알 수 있다 . 이런 점 에서 시각화 는 데이터 분석 에서 아주 중요한요소 라고 할 수 있다 .

Matplotlib 그래프 구성요소

| 구성요소              | 설명                              |
| ----------------- | ------------------------------- |
| **Figure**        | 전체 그래프 영역 (도화지 역할)              |
| **Axes**          | 실제 그래프가 그려지는 영역 (하나의 subplot)   |
| **Axis**          | x축과 y축을 각각 의미 (Ticks, Label 포함) |
| **Title**         | 그래프 제목                          |
| **X/Y Label**     | 각 축의 이름                         |
| **Tick**          | 축 위의 눈금 위치                      |
| **Tick Label**    | 눈금에 붙은 숫자나 텍스트                  |
| **Legend**        | 범례 – 선이나 마커의 의미를 설명             |
| **Grid**          | 격자선 (데이터 해석 도움)                 |
| **Line / Marker** | 실제 데이터 선 또는 마커 점                |


그래프 유형과 활용 목적

| 그래프    | 주요 용도     | 데이터 형태        | 대표 해석                   |
| ------ | --------- | ------------- | ----------------------- |
| 선 그래프  | 추세, 변화량   | 연속형 vs 연속형    | BMI 증가 → Target 증가      |
| 면적 그래프 | 누적/영역 강조  | 연속형           | BMI 위험 영역 시각화           |
| 막대 그래프 | 범주별 평균 비교 | 범주형 vs 수치형    | High 그룹이 평균 Target 가장 큼 |
| 히스토그램  | 분포 분석     | 단일 연속형        | 정규 분포 형태                |
| 산점도    | 관계 확인     | 연속형 vs 연속형    | 양의 상관관계                 |
| 파이차트   | 비율 파악     | 범주형           | 3등분된 BMI 그룹             |
| 박스플롯   | 분포 + 이상치  | 범주형 그룹 vs 수치형 | High 그룹에 이상치 많음         |

plot() 함수 구성요소 설명

| 파라미터       | 설명                                                                                                             |
| ---------- | -------------------------------------------------------------------------------------------------------------- |
| `*args`    | 보통 `x, y` 좌표값 또는 `y`값만 입력<br>예: `plot(x, y)`, 또는 `plot(y)`                                                     |
| `scalex`   | `True`면 x축 자동 스케일 조정 (기본값 `True`)                                                                              |
| `scaley`   | `True`면 y축 자동 스케일 조정 (기본값 `True`)                                                                              |
| `data`     | `dict`, `DataFrame` 등으로 `x`, `y`를 이름으로 전달 가능<br>예: `data={'a':[1,2], 'b':[3,4]}` → `plot('a', 'b', data=data)` |
| `**kwargs` | 색상(color), 선 스타일(linestyle), 마커(marker) 등 추가 옵션                                                                |


자주 사용하는 **kwargs (키워드 인자)

| 옵션 이름               | 예시                           | 설명            |
| ------------------- | ---------------------------- | ------------- |
| `color` or `c`      | `'blue'`, `'r'`, `'#FFAA00'` | 선 색상          |
| `linestyle` or `ls` | `'-'`, `'--'`, `':'`, `'-.'` | 선 스타일         |
| `linewidth` or `lw` | `1.5`                        | 선 굵기          |
| `marker`            | `'o'`, `'s'`, `'x'`, `'^'`   | 각 점에 표시할 마커   |
| `label`             | `'실험값'`                      | 범례(legend) 이름 |
| `alpha`             | `0.5`                        | 투명도 (0 \~ 1)  |





````python
import matplotlib.pyplot as plt

x = [0, 1, 2, 3]
y = [0, 1, 4, 9]
# plot(*args: float | ArrayLike | str, scalex: bool=True, scaley: bool=True, data=None, **kwargs)
plt.plot(x, y)  #plt.plot(y)
plt.title("기본 그래프 그리기")

plt.show()
````

````python
# plot(*args: float | ArrayLike | str, scalex: bool=True, scaley: bool=True, data=None, **kwargs)
plt.plot([1, 2, 3], [2, 4, 6], scalex=False, scaley=False)  # 자동 스케일링 안함 (그래서 선이 안보임)
plt.title("scalex, scaley False 예")
plt.show()
````

````python
# plot(*args: float | ArrayLike | str, scalex: bool=True, scaley: bool=True, data=None, **kwargs)
data = {'year': [2020, 2021, 2022], 'sales': [100, 150, 130]}
plt.plot('year', 'sales', data=data)
plt.xticks(data['year']) # x축을 정수 눈금으로 고정 plt.xticks(range(2020, 2023))
plt.title("data 인자 사용 예")
plt.xlabel("Year")   # x축 이름
plt.ylabel("Sales")  # y축 이름
plt.show()
````

matplotlib의 linestyle 옵션

| 선 스타일          | 설명                       |
| -------------- | ------------------------ |
| `'-'`          | 실선 (solid line)          |

| 부분                                                    | 설명                                                                                      |
| ----------------------------------------------------- | --------------------------------------------------------------------------------------- |
| `<Figure size 640x480 with 4 Axes>`                   | 하나의 Figure(그래프 도화지)를 생성했고, 그 안에 4개의 Axes(서브플롯)가 포함되어 있다는 의미입니다.                         |
| `array([[<Axes: >, <Axes: >], [<Axes: >, <Axes: >]])` | `2행 x 2열` 형태로 구성된 Axes 배열 (즉, 4개의 서브플롯)이 있다는 뜻입니다. 각 `<Axes: >`는 실제 그래프를 그릴 수 있는 영역입니다. |
| `dtype=object`                                        | NumPy 배열의 타입이 객체 타입임을 의미합니다. 각 요소가 `matplotlib.axes._axes.Axes` 객체입니다.                  |

선과 점을 함께 표현(2개의 그래프 표현)

````python

# 선그래프
plt.plot(x, y, linewidth=5, c='r', alpha=0.5, label='line graph')

# 점 그래프
plt.scatter(x, y, s=150, c='g', alpha=0.5, label='dot graph')

#범례
plt.legend(loc=4, fontsize=15)

#ticks는 각 눈금에 표시된 숫자들로 이들의 크기 및 색상을 설정
plt.xticks(fontsize=14, color='b')
plt.yticks(fontsize=14, color='y')

#축에 대한 설명
plt.xlabel('Sun light', fontsize=15, color='b')
plt.ylabel('Apple Sweat', fontsize=15, color='y')


# 수직, 수평 그리기
plt.axvline(x=2, linewidth=10) # 수직축(vertical axis) -> y축 /세로
plt.axhline(y=4, linewidth=8)  # 수평축(horizontal axis) -> x축/ 가로


#제목
plt.title('y=2x', fontsize=20, c='m')
````

````python
import matplotlib.pyplot as plt

fig, ax_dict = plt.subplot_mosaic(
    """
    AB
    CD
    """, figsize=(6, 6)
)

ax_dict['A'].plot([1, 2], [3, 4])
ax_dict['B'].bar([1, 2], [5, 1])
ax_dict['C'].scatter([1, 2, 3], [1, 2, 1])
ax_dict['D'].hist([1, 1, 2, 3])

plt.tight_layout()
plt.show()
````

## 1. 데이터 준비

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_diabetes

# 데이터 로드
diabetes = load_diabetes()
df = pd.DataFrame(diabetes.data, columns=diabetes.feature_names)
df['target'] = diabetes.target
````

## 2. 선 그래프 (Line Plot)

'o': 원 (circle)

's': 사각형 (square)

'^': 삼각형 (triangle)

'D': 다이아몬드 (diamond)

'*': 별 (star)

````python
# bmi 값 순서대로 정렬 후 target 선형관계 확인
sorted_df = df.sort_values(by='bmi')
plt.figure(figsize=(20, 4))
plt.plot(sorted_df['bmi'].values, sorted_df['target'].values, color='blue', linestyle='--', marker='*')
plt.title('선 그래프: BMI vs Target')
plt.xlabel('BMI')
plt.ylabel('Target')
plt.grid(True)
plt.show()

# 사용 데이터: bmi, target
# 용도: 시간 흐름, 순차적 변화, 연속적 관계 확인에 사용
# 해석: BMI가 증가할수록 Target(질병 진행 지표)도 증가하는 경향을 보임 → 양의 상관관계.
````

3. 면적 그래프 (Area Plot)

````python
df_sorted = df.sort_values(by='bmi')
plt.figure(figsize=(8, 4))
plt.fill_between(df_sorted['bmi'], df_sorted['target'], color='lightgreen', alpha=0.5)
plt.title("면적 그래프: BMI vs Target")
plt.xlabel("BMI")
plt.ylabel("Target")
plt.grid(True)
plt.show()

# 사용 데이터: bmi, target
# 용도: 누적된 값, 면적 누적, 변화량 강조
# 해석: BMI 값이 커짐에 따라 Target이 함께 커지는 형태를 면적으로 표현하여 시각적으로 총 위험도 영역을 표현할 수 있음.
````

## 4. 막대 그래프 (Bar Plot)

````python
# bmi를 세 그룹으로 나누고 평균 target 계산
df['bmi_group'] = pd.cut(df['bmi'], bins=3, labels=['Low', 'Medium', 'High'])
grouped = df.groupby('bmi_group')['target'].mean()
print(grouped)

plt.figure(figsize=(6, 4))
grouped.plot(kind='bar', color='skyblue')
plt.title("막대 그래프: BMI Group별 평균 Target")
plt.ylabel("평균 Target")
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.show()

# 사용 데이터: 범주형(bmi_group) vs 수치형(target 평균)
# 용도: 범주 간 비교에 적합
# 해석: BMI가 높을수록 평균 Target도 증가 → BMI가 높은 그룹일수록 질병 위험도 증가
````

## 5. 히스토그램 (Histogram)

````python
plt.figure(figsize=(6, 4))
plt.hist(df['target'], bins=20, color='orange', edgecolor='black')
plt.title("히스토그램: Target 분포")
plt.xlabel("Target")  # x축이 타켓
plt.ylabel("빈도")
plt.grid(True)
plt.show()

# 사용 데이터: 단일 연속형 변수 (target)
# 용도: 분포 형태 파악 (정규성, 치우침 등)
# 해석: 대략 정규분포 형태를 보이며, Target이 100~200 사이에 많이 분포함
````

## 6. 산점도 (Scatter Plot)

````python
plt.figure(figsize=(6, 6))
plt.scatter(df['bmi'], df['target'], alpha=0.6)
plt.title("산점도: BMI vs Target")
plt.xlabel("BMI")
plt.ylabel("Target")
plt.grid(True)
plt.show()

# 사용 데이터: 두 수치형 변수
# 용도: 변수 간 상관관계 파악
# 해석: BMI가 높을수록 Target도 높아짐 → 선형 상관관계 있음
````

## 7. 파이차트 (Pie Chart)

````python
# BMI 그룹별 인원 수
counts = df['bmi_group'].value_counts().sort_index()
plt.figure(figsize=(6, 6))
plt.pie(counts, labels=counts.index, autopct='%1.1f%%', colors=['lightblue', 'lightgreen', 'lightcoral'])
plt.title("파이차트: BMI Group 분포")
plt.show()

# 중요) 사용 데이터: 범주형 변수의 비율 (BMI는 서열형 데이터 (Ordinal Data)에 속함 - Low, Medium, High처럼 순서가 존재하기 때문입니다.)
# 용도: 비율 시각화, 그룹별 크기 비교
# 해석: 세 그룹(Low, Medium, High)이 대체로 균등하게 분포됨
````

## Seaborn

Seaborn v0.14 이후 주요 변경사항 & 대응 가이드

1. palette 사용 시 hue 명시 필수

```
# 이전 방식 (v0.13 이하까지 허용됨)
sns.barplot(data=df, x='bmi_group', y='target', palette='Blues')

# 앞으로 권장 방식 (v0.14 이상)
sns.barplot(data=df, x='bmi_group', y='target', hue='bmi_group',
            palette='Blues', legend=False)

# 이유: palette는 색상 맵핑을 위해 hue가 필요합니다. hue 없이 쓰면 경고 또는 오류가 발생할 수 있습니다.
```

2. ci → errorbar로 변경

```
# 이전
sns.barplot(data=df, x='bmi_group', y='target', ci='sd')

# 현재 (v0.13~)
sns.barplot(data=df, x='bmi_group', y='target', errorbar='sd')

이유: ci(confidence interval)는 더 이상 사용되지 않으며, errorbar='sd', errorbar='se' 등으로 대체되었습니다.
```

# [2] 시본 막대 그래프

errorbar 옵션

| 목적                       | 선택할 errorbar |
| ------------------------ | ------------ |
| 그룹 내 분포의 **산포도**를 보고 싶다면 | `'sd'`       |
| 평균값의 **신뢰도/정확도**를 보고 싶다면 | `'se'`       |
| 오차막대가 필요 없다면             | `None`       |



palette 옵션

| 팔레트 이름                 | 설명              |
| ---------------------- | --------------- |
| `Blues`                | 파란색 계열 밝기 변화    |
| `Reds`                 | 빨간색 계열 밝기 변화    |
| `Greens`               | 초록색 계열 밝기 변화    |
| `Purples`              | 보라색 계열 밝기 변화    |
| `Set1`, `Set2`, `Set3` | 대비가 강한 여러 색상 세트 |
| `coolwarm`             | 차가운 색과 따뜻한 색 조합 |
| `viridis`              | 밝기 순서가 명확한 컬러맵  |

# [8] 히트맵 (상관계수 시각화)

````python
import numpy as np

# 숫자 타입 컬럼만 선택해 상관계수 계산
numeric_df = df.select_dtypes(include=[np.number])
corr = numeric_df.corr()  # 수치형 컬럼 간 피어슨 상관계수 계산

plt.figure(figsize=(10, 6))

# 상관계수 행렬을 heatmap으로 시각화
# annot=True: 각 셀에 상관계수 숫자 표시
# cmap='coolwarm': 양수는 붉은색 계열, 음수는 파란색 계열로 색 구분
# fmt=".2f": 소수점 둘째 자리까지 숫자 표시
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")

plt.title("[시본] 피처 간 상관관계 히트맵")
plt.show()


# 실행 결과 설명
# 사용 데이터: 데이터프레임 내 모든 수치형 변수들

# 용도: 변수들 간 상관관계 탐색 및 변수 간 선형관계 강도 파악

# 해석:

# +1에 가까운 값은 강한 양의 상관관계, -1에 가까운 값은 강한 음의 상관관계를 의미

# 예를 들어, bmi와 target의 상관계수가 0.59 정도라면, BMI가 높을수록 당뇨 진행도(target)가 증가하는 경향이 있음을 시사

# 강한 상관관계 변수 쌍은 모델링 시 중요 변수 후보로 고려 가능

# 상관관계가 약하거나 0에 가까운 변수들은 독립적이라고 볼 수 있음
````

# [9] 다중 플롯 (pairplot)

````python
# diag_kind='kde': 대각선에 각 변수의 커널 밀도 추정(KDE) 그래프를 표시 (히스토그램 대신 부드러운 분포 곡선)
sns.pairplot(df[['bmi', 'bp', 's5', 'target']], diag_kind='kde')

# pairplot은 자체적으로 Figure를 생성하므로 plt.suptitle()로 제목을 추가할 때 y 위치를 약간 위로 조정
plt.suptitle("[시본] 주요 변수 관계 시각화", y=1.02)

plt.show()


# 실행 결과 설명
# 사용 데이터: BMI, 혈압(bp), s5, 당뇨 진행도(target) 주요 수치형 변수

# 용도: 변수 간 쌍별 관계 탐색과 분포 형태 시각화

# 그래프 구성:

#   ①대각선: 각 변수별 KDE 분포곡선
#   ②비대각선: 변수 쌍별 산점도(Scatter plot)

# 해석:

#  ①산점도에서 점들이 일정한 선형 패턴을 보이면 양/음의 상관관계가 있음을 시사
#  ②BMI와 target 간에는 양의 상관관계가 보여 BMI가 클수록 당뇨 진행도가 높을 가능성 암시
#  ③대각선 KDE를 통해 각 변수의 분포 형태(정규성, 치우침 등)를 확인할 수 있음
#   ④변수 쌍 간 관계를 한눈에 파악할 수 있어 데이터 탐색(EDA)에 유용
````
