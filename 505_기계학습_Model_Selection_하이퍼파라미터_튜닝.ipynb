### êµì°¨ê²€ì¦(3way holdout)

ë°ì´í„°ë¥¼ ì´ 3ë“±ë¶„í•œë‹¤.
- train set : ëª¨ë¸ í•™ìŠµì— ì‚¬ìš©í•œë‹¤.
- validationt set : í•˜ì´í¼ íŒŒë¼ë¯¸í„°ë¥¼ ì°¾ê¸° ìœ„í•´ì„œ ì‚¬ìš©í•œë‹¤.
- test set : í•œë²ˆë„ ì•ˆ ë³¸ ë°ì´í„°ë¡œ ëª¨ë¸ì˜ ì„±ëŠ¥ì„ ì¸¡ì •í•˜ê¸° ìœ„í•´ì„œ ì‚¬ìš©í•œë‹¤.

````python
X_train, X_test, y_train, y_test = train_test_split(iris.iloc[:, :-1], iris.target)
````

````python
X_train, X_validation, y_train, y_validation = train_test_split(X_train, y_train)
````

 **"í•˜ì´í¼íŒŒë¼ë¯¸í„° íŠœë‹"**ì„ ìœ„í•œ 3-way holdout ë°©ì‹ì„ ì‚¬ìš©í•´ì„œ ìµœì ì˜ K ê°’ì„ ì°¾ëŠ” ê³¼ì •

K=1

````python
# í•˜ì´í¼ íŒŒë¼ë¯¸í„°ë¥¼ ì°¾ê¸° ìœ„í•´ì„œ 3-way holdoutì„ í•œë‹¤.
temp = []
for i in range(1, 20):
    knn = KNeighborsClassifier(n_neighbors=i) # ì´ì›ƒì˜ ìˆ˜Kë¥¼ ë³´ê³  ê²€ì‚¬ë¥¼ í•œë‹¤ K=1 (ì´ì›ƒ1ê°œë§Œ ë³´ê³  ê²€ì‚¬í•œë‹¤)
    knn.fit(X_train, y_train)
    temp.append(knn.score(X_validation, y_validation))
````
